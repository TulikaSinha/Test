### Nested For

for [first iterating variable] in [outer loop]: # Outer loop
    [do something]  # Optional
    for [second iterating variable] in [nested loop]:   # Nested loop
        [do something]  

```python
adj = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for x in adj:
  for y in fruits:
    print(x, y)
```

```python
num_list = [1, 2, 3]
alpha_list = ['a', 'b', 'c']

for number in num_list:
    print(number)
    for letter in alpha_list:
        print(letter)
```

```python
list_of_lists = [['great', 'learning', 'datascience'],[0, 1, 2],[9.9, 8.8, 7.7]]

for list in list_of_lists:
    for item in list:
        print(item)
```

```python
#Find factorial of the number 6

## 6! FACTORIAL is denoted by 1*2*3*4*5*6 = 720

#num = 6

num = int(input("Enter a number: "))

factorial = 1

# check if the number is negative, positive or zero
if num < 0:
   print("Sorry, factorial does not exist for negative numbers")
elif num == 0:
   print("The factorial of 0 is 1")
else:
   for i in range(1,num + 1):
       factorial = factorial*i
print("The factorial of",num,"is",factorial)
```

### Break

```python
# once it checks for the condition and if its satisfied, ignore all the character after that
for i in "greatlearning":
    if i == "i":
        break
    print(i)

print("The end of code")
```

## Continue

```python
# once it checks for the condition and if its satisfied, ignore the current element and prints the rest
for i in "greatlearning":
    if i == "i":
        continue
    print(i)

print("The end of code")
```

```python
# for a given range of values 110,130. Print a message id 124 is found and print all the other values 
# if and continue statement

# for a given range of values b/w 110 to 130 run a loop and print "match is found" if id/value = 124 and then continue the loop.

for x in range(110,130): 
    print("value of x :", x)
    if(x == 124):
        print("There is match found")
        continue #continue statement checks for the condition and continues the loop
```

```python
# Cumulative Sum

n = int(input("enter the input number :"))

# initialize sum and counter
sum = 0
i = 1

while i <= n:
    sum = sum + i
    i = i+1    # update counter

    # print the sum
    print("The sum is", sum)
    
#print("The sum is", sum)
```

### While Break and continue

```python
n = 5
while n > 0:
    n -= 1
    if n == 2:
        break
    print(n)
print('Loop ended.')
```

```python
n = 5
while n > 0:
    n -= 1
    if n == 2:
        continue
    print(n)
print('Loop ended.')
```

### While Loop with else

```python
# initialize  counter
counter = 0

while counter < 3:
    print("Inside loop")
    counter = counter + 1
else:
    print("Inside else") # as the count reached 3, it exits out of the loop
```

```python
# Prints out 0,1,2,3,4 and then it prints "count value reached 5"

count=0
while(count<5):
    print(count)
    count +=1
else:
    print("count value reached %d" %(count))
```

### While(1) - It is an infinite loop which will run till a break statement is issued explicitly

```python
# Prints out 0,1,2,3,4

count = 0
while(1): # while True:
    print(count)
    count += 1
    if count >= 5:
        break
```

#### 5. Write a program to print prime numbers between 2 to 50 ?

```python
noprimes = [j for i in range(2, 8) for j in range(i*2, 50, i)] 
primes = [x for x in range(2, 50) if x not in noprimes] 
primes 
```

##### 6.Write a Program which converts all the letters in the list to UpperCase letters.

```python
letters = ['a', 'b', 'c', 'd']
print(letters)

upper_letters = []
for letter in letters:
  result = letter.upper()
  upper_letters.append(result)

print(upper_letters)
```

#### Same Problem in List comprehension?

```python
letters = ['a', 'b', 'c', 'd']
print(letters)

upper_letters = [x.upper() for x in letters]

print(upper_letters)
```
