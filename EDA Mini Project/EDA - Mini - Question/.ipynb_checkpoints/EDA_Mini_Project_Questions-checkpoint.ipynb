```python
#importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Created the Dataset
df_income = pd.DataFrame({'Hourly_Income':[1000,2009,24418,444478,324235,243242,3434234,7567457,9235,238237,1312,3412],
                          'Hourly_Expense':[651361,217371,2746,2356,13436,5732,346346,3463,1132,23534,242235,235235],
                          'family_members_count':[3,4,2,3,1,4,5,6,3,6,3,5],
                          'House_rent':[1299,2300,3411,3422,4566,4211,4600,736,672,0,734,2374],
                          'Highest_income_Member':["Olivia","George","Isla","Harry","Ava","Noah","Sophia","Jacobi","Freddie","Ella","Grace","Ella"]})
df_income
```

# A. Display the five point summary of the data.

```python
print("Max is : ",df_income.max())
print()
print("Min is : ",df_income.min())
print()
print("Q1 : ",df_income.quantile(0.25))
print()
print("Q2 : ",df_income.quantile(0.5))
print()
print("Q3 : ",df_income.quantile(0.75))
```

# E. Calculate IQR(the difference between 75% and 25% quartile) for Hourly_Income and Hourly_Expense.

```python
q1 = df_income['Hourly_Income'].quantile(0.25)
print("25% Quartile of Hourly_Income is : ",q1)
```

```python
q3 = df_income['Hourly_Income'].quantile(0.75)
print("75% Quartile of Hourly_Income is : ",q3)
```

```python
iqr = q3-q1
print("IQR is 75% - 25% for Hourly_income : ",iqr)
```

```python
q1e = df_income['Hourly_Expense'].quantile(0.25)
print("25% Quartile of Hourly_Expense is : ",q1)
q3e = df_income['Hourly_Expense'].quantile(0.75)
print("75% Quartile of Hourly_Expense is : ",q3)
iqre = q3e-q1e
print("IQR is 75% - 25% for Hourly_Expense : ",iqre)
```

```python
#Step 1: Create a  condition = 'LoanStatus'== 'Completed'| 'LoanStatus' == 'Defaulted' |\'LoanStatus' == 'Chargedoff'
df_condition = df_priLoad[(df_priLoad['LoanStatus']=='Completed') | (df_priLoad['LoanStatus']=='Defaulted') | (df_priLoad['LoanStatus']=='Chargedoff')]
df_condition.head()
```

```python
#Step2: Create a user define function  using condition and `LoanStatus` column.
def funcCondition(x):
    if x=='Completed':
        return 1
    elif x=='Defaulted':
        return 1
    else:
        return 0
```

```python
df_condition['Loan_Status_New']=df_condition['LoanStatus'].apply(funcCondition)
x = df_condition.groupby(by=df_condition['LoanStatus'])['Loan_Status_New'].count()
x[[2]]=x[0]+x[2]
pd.DataFrame(x[[1,2]])

#df_condition['LoanStatus'].value_counts()
```

```python
#B. Step 1: Create a  dictionary called categories = 1: 'Debt Consolidation', 2: 'Home Improvement', 3: 'Business', 6: 'Auto', 7: 'Other'
categories = {1: 'Debt Consolidation', 2: 'Home Improvement', 3: 'Business', 6: 'Auto', 7: 'Other'}
```

```python
#Step2: Create a user define function  using categories and `ListingCategory (numeric)` column.
def listing(x):
    d={1: 'Debt Consolidation', 2: 'Home Improvement', 3: 'Business', 6: 'Auto', 7: 'Other',8: 'Other',9: 'Other',10: 'Other'
       ,11:'Other',12:'Other',13:'Other',14:'Other',15:'Other',16:'Other',17:'Other',18:'Other',19:'Other',20:'Other'}
    if x in d:
        return d[x]
    
df_priLoad['ListingCategory (numeric) new']=df_priLoad['ListingCategory (numeric)'].apply(listing)
```

```python
y=df_priLoad.groupby(by=df_priLoad['ListingCategory (numeric) new']).count()
pd.DataFrame(y['ClosedDate'])
```
